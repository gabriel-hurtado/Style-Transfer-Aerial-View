```python
from PIL import Image, ImageFile
import os
import sys
from os import listdir
from os.path import isfile, join
ImageFile.LOAD_TRUNCATED_IMAGES = True

path="/mnt/agp/data/raw/_ortho2015_png/"
onlyfiles = [f for f in listdir(path) if isfile(join(path, f))]
sizes=[1000]

for ind,fileN in enumerate(onlyfiles) :
    pilImg = Image.open( path+fileN)
    width, height = pilImg.size
    new_im = Image.new('RGB', (width,height))
    for size in sizes:
        
        numSub=width/size
        for i in range (0,numSub):
            for j in range(0,numSub):
                try:
                    img2 = pilImg.crop((i*size,j*size, i*size+size, j*size+size))
                    img2.save("/home/GTL/ghurtado/DeepLearning/adain/fast-style-transfer/tmp/_styled.jpg", 'jpeg')
                    print("stylyzing")
                    #we call the python script for each piece of the image, and  generate a stylized version of it to be pasted in the big reconstruction
                    #   arguments are: script name (do not change) ; 'checkpoint_dir','dir or .ckpt file to load checkpoint from';'in_path', path were the picture to stylize is; 'out_path' were to save
                    sys.argv = ['evaluate.py','/home/GTL/ghurtado/DeepLearning/adain/fast-style-transfer/modelNew', '/home/GTL/ghurtado/DeepLearning/adain/fast-style-transfer/tmp/','/home/GTL/ghurtado/DeepLearning/adain/fast-style-transfer/tmpOut/']
                    execfile('evaluate.py')
                    img2= Image.open( "/home/GTL/ghurtado/DeepLearning/adain/fast-style-transfer/tmpOut/_styled.jpg")
                    new_im.paste(img2, (i*size,j*size, i*size+size, j*size+size))
                except IOError:
                    print("error cropping") 
                
                    


                
    new_im.save("/home/GTL/ghurtado/DeepLearning/adain/fast-style-transfer/res/"+fileN+"_styled.jpg", 'jpeg')

        
```
